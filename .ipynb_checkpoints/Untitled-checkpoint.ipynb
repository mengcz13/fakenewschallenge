```python
%matplotlib inline

import os
import sys
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
# 绘图函数

def draw_comparison(xlabel, ylabel, title, recpairs):
    plt.figure()
    
    fig, ax = plt.subplots()
    for rec, name in recpairs:
        ax.plot(rec, label=name)
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
    
xlabel = 'epoch'
ylabel = 'accuracy'
title = 'Comparison'
recpairs = [([1,2,3,4], 'm1'), ([2,3,4,5], 'm222')]
draw_comparison(xlabel, ylabel, title, recpairs)
```

```python
# 选择n个模型文件夹, 打印loss, accuracy, score的对比图
def compare_models(modelflist, mroot='./savedmodels', recname='records.pkl'):
    recs = []
    for modelf in modelflist:
        recp = os.path.join(mroot, modelf, recname)
        with open(recp, 'rb') as f:
            recs.append((pickle.load(f), str(modelf)))

    trains = [([r1['train_result'] for r1 in r[0]], r[1]) for r in recs]
    valis = [([r1['validate_result'] for r1 in r[0]], r[1]) for r in recs]
    sets = [('Training', trains), ('Validation', valis)]
    itemnames = ['loss', 'accuracy', 'score']
    itemranks = [0, 1, 3]
    for setname, setn in sets:
        for itemname, itemrank in zip(itemnames, itemranks):
            recpairs = [([r1[itemrank] for r1 in r[0]], r[1]) for r in setn]
            draw_comparison('epoch', itemname, ' '.join((setname, itemname)), recpairs)
            
compare_models(os.listdir('./savedmodels/'))
```
